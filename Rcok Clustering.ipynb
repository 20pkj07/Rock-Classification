### Q.1  K-means Clustering algorithm

```python
#importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import numpy as np
```

```python
#loading data
df=pd.read_csv("logdata-3.dat")
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#plotting scatter plot
sns.set(font_scale=1.5)
plt.figure(figsize=(20,20))
sns.pairplot(df[:-1])
plt.show()
```

```python
#corelation
plt.figure(figsize=(10,8))
sns.heatmap(data=df.corr())
```

```python
# Creating the KMeans object
# 3 cluster
km3 = KMeans(n_clusters=3, max_iter=15, tol=0.0001)

# Fitting the KMeans model to the data
km3.fit(df)

# Printting the cluster centers and error for 15 iteration
print("\nCenters   :\n\n",km3.cluster_centers_)
print("\n Error  : ",km3.inertia_)
print('\n')
```

```python
#predicting data
y=km3.fit_predict(df)
y
```

```python
df2=df
df2['Target']=y
df2
```

```python
#plotting scatter plot
plt.figure(figsize=(20,20))
sns.pairplot(df2[:-1],hue='Target',palette='Accent')
plt.show()
```

```python
#counting no. of target
df2['Target'].value_counts()
```

```python
# Creating the KMeans object
# 2 cluster
km2 = KMeans(n_clusters=2, max_iter=15, tol=0.0001)

# Fitting the KMeans model to the data
km2.fit(df)

# Printting the cluster centers and error for 15 iteration
print("\nCenters   :\n\n",km2.cluster_centers_)
print("\n Error  : ",km2.inertia_)
print('\n')
```

```python
x=km2.fit_predict(df)
df3=df
df3['Target']=x
df3
```

```python
#plotting scatter plot
plt.figure(figsize=(20,20))
sns.pairplot(df3[:-1],hue='Target',palette='RdBu')
plt.show()
```

```python
#counting no. of target
df3['Target'].value_counts()
```
