```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Test data generation

```python

df2=df
pred=[]
for i in range(1408):

    if df['density(g/cc)'][i]>=2.64 and df['density(g/cc)'][i]<=2.77 and df['porosity(%)'][i]>=9 and df['porosity(%)'][i]<=10 and df['Gamma(API)'][i]>=109 and df['Gamma(API)'][i]<=127 and df['sonic(ms/m'][i]>=167 and df['sonic(ms/m'][i]<=176 and df['resistivity(ohm-m)'][i]>=7.5 and df['resistivity(ohm-m)'][i]<=8.83:
        pred.append(1)
    elif df['density(g/cc)'][i]>=2.90 and df['density(g/cc)'][i]<=3.30 and df['porosity(%)'][i]>=17 and df['porosity(%)'][i]<=26 and df['Gamma(API)'][i]>=4 and df['Gamma(API)'][i]<=39 and df['sonic(ms/m'][i]>=148 and df['sonic(ms/m'][i]<=160 and df['resistivity(ohm-m)'][i]>=8.0 and df['resistivity(ohm-m)'][i]<=9.0:
        pred.append(2)
    elif df['density(g/cc)'][i]>=1.90 and df['density(g/cc)'][i]<=2.80 and df['porosity(%)'][i]>=15 and df['porosity(%)'][i]<=19 and df['Gamma(API)'][i]>=57 and df['Gamma(API)'][i]<=73 and df['sonic(ms/m'][i]>=162 and df['sonic(ms/m'][i]<=171 and df['resistivity(ohm-m)'][i]>=7.0 and df['resistivity(ohm-m)'][i]<=8.0:
        pred.append(3)
    else :
        pred.append(0)
df2['Class']=pred
```

```python
df2
```

```python
df2['Class'].value_counts()
```

```python
X=df2.drop('Class',axis=1)
X
```

```python
y=df2.Class
y
```

# Training and testing model on original data

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,stratify=y)
```

```python
# Training model on original data
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow import keras

# Defineing the model architecture
model = Sequential()
model.add(Dense(10, input_dim=5, activation='relu'))
model.add(Dense(4, activation='softmax'))

# Compileing the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

history=model.fit(X_train,y_train,epochs=200,validation_data=(X_test,y_test))
```

```python
train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
plt.figure(figsize=(16,9))
epochs = range(1,201)
plt.plot(epochs, train_acc, 'g', label='Training accuracy')
plt.plot(epochs, val_acc , 'b', label='validation accuracy')
plt.title('Training and Validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Printing accuracy of model
y_pred=model.predict(X_test)
y_pre = np.argmax(y_pred, axis=1)
print("\n")
print("Accuracy       : ",accuracy_score(y_test,y_pre))
print("Report         : \n",classification_report(y_test,y_pre))
print(pd.crosstab(y_test,y_pre))
print('\n')
```
